```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def get_saturation(frame):
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_RGB2HSV)
    return np.mean(frame_hsv[:,:,1])

def augment_saturation(frame, saturation_factor):
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_RGB2HSV)
    frame_hsv[:,:,1] = np.clip(frame_hsv[:,:,1] * saturation_factor, 0, 255).astype(np.uint8)
    return cv2.cvtColor(frame_hsv, cv2.COLOR_HSV2RGB)

original_saturation_per_video = []
augmented_saturation_per_video = []

# Custom weights for different ranges of saturation
saturation_choices = np.array([i for i in range(20, 141)])
weights = np.array([3 if i <= 60 else 1 for i in range(20, 141)])

# Normalize weights
weights = weights / np.sum(weights)

# Loop over all subjects in the root folder
for subject_folder in os.listdir('/Users/qpingwin/labs/KlarityAII/full_UBFC'):
    
    subject_path = os.path.join('/Users/qpingwin/labs/KlarityAII/full_UBFC', subject_folder)

    if os.path.isdir(subject_path):
        
        for file in os.listdir(subject_path):
            if file.endswith('.avi'):
                video_path = os.path.join(subject_path, file)
                cap = cv2.VideoCapture(video_path)
                frame_saturation_values = []

                while cap.isOpened():
                    ret, frame = cap.read()
                    if not ret:
                        break

                    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                    frame_saturation_values.append(get_saturation(frame_rgb))

                # Compute original average saturation for this video
                avg_original_saturation = np.mean(frame_saturation_values)
                original_saturation_per_video.append(avg_original_saturation)

                # Compute saturation factor to reach target saturation
                target_saturation = np.random.choice(saturation_choices, p=weights)
                saturation_factor = target_saturation / avg_original_saturation
                
                # Reset capture to augment frames with consistent saturation
                cap = cv2.VideoCapture(video_path)
                frame_saturation_values = []
                
                while cap.isOpened():
                    ret, frame = cap.read()
                    if not ret:
                        break

                    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                    augmented_frame = augment_saturation(frame_rgb, saturation_factor)
                    frame_saturation_values.append(get_saturation(augmented_frame))

                # Compute augmented average saturation for this video
                avg_augmented_saturation = np.mean(frame_saturation_values)
                augmented_saturation_per_video.append(avg_augmented_saturation)

                cap.release()
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate average saturation for a frame
def get_saturation(frame):
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    return np.mean(frame_hsv[:, :, 1])

video_mean_original_saturation_list = []

# Loop over all main folders (e.g., 01-01, 01-02)
for main_folder in os.listdir('/Users/qpingwin/labs/KlarityAII/Pure_dataset'):
    if main_folder == 'pure':
        continue  # Skip this iteration and move to the next folder
    main_path = os.path.join('/Users/qpingwin/labs/KlarityAII/Pure_dataset', main_folder)
    
    if os.path.isdir(main_path):
        for sub_folder in os.listdir(main_path):
            sub_path = os.path.join(main_path, sub_folder)
            
            if os.path.isdir(sub_path):
                original_saturation_values = []
                
                for file in os.listdir(sub_path):
                    if file.endswith('.png'):
                        frame_path = os.path.join(sub_path, file)
                        frame = cv2.imread(frame_path)
                        
                        original_saturation = get_saturation(frame)
                        original_saturation_values.append(original_saturation)
                
                avg_original_saturation = np.mean(original_saturation_values)
                video_mean_original_saturation_list.append(avg_original_saturation)



# Create histogram
plt.figure(figsize=(12, 6))

plt.hist(video_mean_original_saturation_list, bins=10, alpha=0.7, label='Original', density=True)
plt.title('Histogram of Original Mean Saturation in Pure Dataset')
plt.xlabel('Mean Saturation per Video')
plt.ylabel('Frequency Density')
plt.legend()

plt.show()

```
